```python
"""
Definition of ListNode
class ListNode(object):
    def __init__(self, val, next=None):
        self.val = val
        self.next = next
"""

class Solution:
    """
    @param headA: the first list
    @param headB: the second list
    @return: a ListNode
    """
    def getIntersectionNode(self, headA, headB):
        # deal with the corner case
        if not headA or not headB:
            return 
        
        len_A, len_B = 0, 0
        nodeA, nodeB = headA, headB
        
        # get the length of headA and headB
        while nodeA != None:
            len_A += 1
            nodeA = nodeA.next

        while nodeB != None:
            len_B += 1
            nodeB = nodeB.next        
        
        # calculate the length difference
        dif_len = abs(len_A - len_B)
        
        # find the intersection
        if len_A > len_B:
            for i in range(dif_len):
                headA = headA.next
            
            while headA != headB:
                headA = headA.next
                headB = headB.next
            
            return headA
            
        if len_B > len_A:
            for i in range(dif_len):
                headB = headB.next
            
            while headA != headB:
                headA = headA.next
                headB = headB.next
            
            return headB
            
        if len_A == len_B:
            while headA != headB:
                headA = headA.next
                headB = headB.next
            
            return headA
```
