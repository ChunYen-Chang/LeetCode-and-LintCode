```python
"""
Definition of TreeNode:
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left, self.right = None, None
"""

class Solution:
    """
    @param root: the given BST
    @param target: the given target
    @param k: the given k
    @return: k values in the BST that are closest to the target
    """
    def inorder_traversal(self, root, result):
        if not root:
            return
        
        self.inorder_traversal(root.left, result)
        result.append(root.val)
        self.inorder_traversal(root.right, result)
        
        
    def binary_search(self, array, target):
        start = 0
        end = len(array)-1
        
        while start+1 < end:
            mid = (start+end) // 2
            if array[mid] > target:
                end = mid
            elif array[mid] <= target:
                start = mid
        return start, end
        
    
    def find_close_element(self, traversal_result, start, end, k, target):
        if len(traversal_result) == 1:
            return traversal_result
            
        left = start
        right = end
        closet_elements = []
    
        while len(closet_elements) < k:
            if end == len(traversal_result):
                closet_elements.append(traversal_result[start])
                start -= 1
            elif start < 0:
                closet_elements.append(traversal_result[end])
                end += 1   
            else:
                if abs(target - traversal_result[start]) > abs(target - traversal_result[end]):
                    closet_elements.append(traversal_result[end])
                    end += 1
                else:
                    closet_elements.append(traversal_result[start])
                    start -= 1
        return closet_elements
        
        
    def closestKValues(self, root, target, k):
        traversal_result = []
        self.inorder_traversal(root, traversal_result)        
        start, end = self.binary_search(traversal_result, target)
        closet_elements = self.find_close_element(traversal_result, start, end, k, target)
        
        return closet_elements

```
