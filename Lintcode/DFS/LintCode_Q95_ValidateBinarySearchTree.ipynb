```python
# ANSWER ONE - using inorder traversal
```

```python
"""
Definition of TreeNode:
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left, self.right = None, None
"""

class Solution:
    """
    @param root: The root of binary tree.
    @return: True if the binary tree is BST, or false
    """
    def __init__(self):
        self.last_value = None
        self.isBST = True
    
    def traversal(self, root):
        if not root:
            return

        self.traversal(root.left)
        if self.last_value != None and root.val <= self.last_value:
            self.isBST = False
            return 
        self.last_value = root.val
        self.traversal(root.right)
    
    
    def isValidBST(self, root):
        self.traversal(root)
        return self.isBST
```

```python
"""
Definition of TreeNode:
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left, self.right = None, None
"""

class Solution:
    """
    @param root: The root of binary tree.
    @return: True if the binary tree is BST, or false
    """
    def BFS(self, root):
        queue = []
        queue.append(root)
        
        while queue:
            level_size = len(queue)
            previous_num = None
            
            for i in range(level_size):
                current_node = queue.pop(0)
                
                if not previous_num:
                    previous_num = current_node.val
                elif current_node.val <= previous_num:
                    return False
                elif current_node.val > previous_num:    
                    previous_num = current_node.val
        return True
    
    def isValidBST(self, root):
        if not root:
            return True
        
        return self.BFS(root)
```

```python
"""
Definition of TreeNode:
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left, self.right = None, None
"""

class Solution:
    """
    @param root: The root of binary tree.
    @return: True if the binary tree is BST, or false
    """
    def divide_and_conquer(self, root):
        if not root:
            return True, None, None
            
        left_is_bst, left_min, left_max = self.divide_and_conquer(root.left)
        right_is_bst, right_min, right_max = self.divide_and_conquer(root.right)

        if not left_is_bst or not right_is_bst:
            return False, None, None
        
        if left_max and left_max >= root.val:
            return False, None, None
            
        if right_min and right_min <= root.val:
            return False, None, None

        if left_min:
            min_node = left_min
        else:
            min_node = root.val
        
        if right_max:
            max_node = right_max
        else:
            max_node = root.val


        return True, min_node, max_node
    
    def isValidBST(self, root):
        if not root:
            return True
        root_is_bst, min_node, max_node = self.divide_and_conquer(root)
        return root_is_bst
```
