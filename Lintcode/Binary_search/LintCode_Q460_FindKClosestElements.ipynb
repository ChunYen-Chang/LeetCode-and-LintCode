```python
class Solution:
    """
    @param A: an integer array
    @param target: An integer
    @param k: An integer
    @return: an integer array
    """
    def two_points_target_find(self, point, A, k):
        res = [A[point]]
        left = point - 1
        right = point + 1

        while len(res) < k:        
            if (left >= 0) and (right <= len(A)-1):
                left_diff = abs(A[left] - A[point])
                right_diff = abs(A[right] - A[point])        
                if left_diff <= right_diff:
                    res.append(A[left])
                    left -= 1
                elif right_diff < left_diff:
                    res.append(A[right])
                    right += 1
                
            elif (left < 0) and (right <= len(A)-1):
                res.append(A[right])
                right += 1
        
            elif (left >= 0) and (right > len(A)-1):
                res.append(A[left])
                left -= 1          
        return res

            
    def two_points_target_not_find(self, target_value, left_point, right_point, A, k):    
        res = []   
        left = left_point
        right = right_point
    
        while len(res) < k:        
            if (left >= 0) and (right <= len(A)-1):
                left_diff = abs(A[left] - target_value)
                right_diff = abs(A[right] - target_value)        
                if left_diff <= right_diff:
                    res.append(A[left])
                    left -= 1
                elif right_diff < left_diff:
                    res.append(A[right])
                    right += 1
                
            elif (left < 0) and (right <= len(A)-1):
                res.append(A[right])
                right += 1
        
            elif (left >= 0) and (right > len(A)-1):
                res.append(A[left])
                left -= 1          
        return res        
        
    
    def kClosestNumbers(self, A, target, k):
        if k == 0:
            return []
            
        start = 0
        end = len(A) - 1
        
        while start+1 < end:
            mid = (start + end) // 2
            if A[mid] >= target:
                end = mid
            elif A[mid] < target:
                start = mid
                
        if A[start] == target:
            return self.two_points_target_find(start, A, k)
            
        if A[end] == target:
            return self.two_points_target_find(end, A, k)
            
        return self.two_points_target_not_find(target, start, end, A, k)
```

```python
# note: the basic idea for solving this problem is "binary search + two pointers"
#       1. when we design the two pointer functions, we have to consider the left and right side index 
#          might go beyond the list range. To tackle this, we give the if statement in these two functions
#       2. when we design the two pointer, please do not think about moving two pointers at the same time
#          the correct idea is we move one, make a comparison, and move next one
```
