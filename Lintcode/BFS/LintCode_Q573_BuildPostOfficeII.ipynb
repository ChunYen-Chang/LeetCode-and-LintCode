```python
class Solution:
    """
    @param grid: a 2D grid
    @return: An integer
    """
    def count_buildings(self, grid):
        num_buildings = 0
        for row in range(len(grid)):
            for column in range(len(grid[0])):
                if grid[row][column] == 1:
                    num_buildings += 1
        return num_buildings
        
        
    def get_empty_location(self, grid):
        empty_location = []
        for row in range(len(grid)):
            for column in range(len(grid[0])):
                if grid[row][column] == 0:
                    empty_location.append([row, column])
        return empty_location    
    
    
    def is_valid(self, grid, new_row, new_column, visited):
        length_row = len(grid)
        length_column = len(grid[0])
        
        if (0 <= new_row < length_row) and (0 <= new_column < length_column):
            if grid[new_row][new_column] != 2:
                if [new_row, new_column] not in visited:
                    return True
        else:
            return False
    
    
    def bfs(self, grid, empty_location, directions, total_building_num):
        queue = []
        queue.append(empty_location)
        building_num = 0
        distance = 0
        sum_distance = 0
        visited = []
        
        while queue:
            distance += 1
            for i in range(len(queue)):
                current_location = queue.pop(0)
                visited.append(current_location)
            
                for move_row, move_column in directions:
                    new_row = current_location[0] + move_row
                    new_column = current_location[1] + move_column
                    if self.is_valid(grid, new_row, new_column, visited):
                        if grid[new_row][new_column] == 1:
                            building_num += 1
                            sum_distance += distance
                            if building_num == total_building_num:
                                return sum_distance
                        queue.append([new_row, new_column])                
        return -1
        
    
    def shortestDistance(self, grid):
        directions = [[1, 0], [0, 1], [-1, 0], [0, -1]]
        total_building_num = self.count_buildings(grid)
        empty_location = self.get_empty_location(grid)
        
        distance_sum_list = []
        for each_empty_location in empty_location:
            distance_sum = self.bfs(grid, each_empty_location, directions, total_building_num)
            distance_sum_list.append(distance_sum)
        
        # make sure we can decide a place to have the post office(can find the min value)
        if len(set(distance_sum_list)) == 1:
            return -1
        else:
            return min(distance_sum_list)
```
