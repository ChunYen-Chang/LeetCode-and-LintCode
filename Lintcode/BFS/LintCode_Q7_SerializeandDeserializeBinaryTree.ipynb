```python
"""
Definition of TreeNode:
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left, self.right = None, None
"""


class Solution:
    """
    @param root: An object of TreeNode, denote the root of the binary tree.
    This method will be invoked first, you should design your own algorithm 
    to serialize a binary tree which denote by a root node to a string which
    can be easily deserialized by your own "deserialize" method later.
    """
    def serialize(self, root):
        def get_tree_element(root):
            queue = []
            tree_element = []
            queue.append(root)
        
            while queue:
                current_node = queue.pop(0)
            
                if not current_node:
                    tree_element.append("#")
                else:
                    tree_element.append(str(current_node.val))
                    queue.append(current_node.left)
                    queue.append(current_node.right)
        
            return tree_element
    
        def drop_duplicate_hash(alist):
            while alist[-1] == "#":
                alist.pop()
            return alist
            
        def convert_list_to_str(alist):
            return "{" + ",".join(alist) + "}"
    
        # Start the serialization process
        if not root:
            return "{}"
        elements = get_tree_element(root)
        result_list = drop_duplicate_hash(elements)
        result_list = convert_list_to_str(result_list)
    
        return result_list

    
    """
    @param data: A string serialized by your serialize method.
    This method will be invoked second, the argument data is what exactly
    you serialized at method "serialize", that means the data is not given by
    system, it's given by your own serialize method. So the format of data is
    designed by yourself, and deserialize it here as you serialize it in 
    "serialize" method.
    """
    def deserialize(self, data):
        def str_to_list(data):
            data = data.strip("{").strip("}")
            data_list_format = data.split(",")
            return data_list_format
    
        def list_to_tree(alist):
            queue = []
            root = TreeNode(int(alist[0]))
            queue.append(root)
        
            for i in range(1, len(alist), 2):
                current_node = queue.pop(0)

                if alist[i] != "#":
                    current_node.left = TreeNode(int(alist[i]))
                    queue.append(current_node.left)
                
                if alist[i+1] != "#":    
                    current_node.right = TreeNode(int(alist[i+1]))
                    queue.append(current_node.right)
            
            return root
    
        # start the deserialization process
        if data == "" or data == "{}":
            return {}
    
        data_list = str_to_list(data)
        data_tree = list_to_tree(data_list)
    
        return data_tree
```
