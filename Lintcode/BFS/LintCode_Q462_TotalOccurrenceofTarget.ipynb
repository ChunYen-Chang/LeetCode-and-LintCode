```python
class Solution:
    """
    @param A: A an integer array sorted in ascending order
    @param target: An integer
    @return: An integer
    """
    def binary_search_front(self, A, target):
        if A == []:
            return -1
            
        start = 0
        end = len(A)-1
        
        while start+1 < end:
            mid = (start+end) // 2
            
            if A[mid] >= target:
                end = mid
            elif A[mid] < target:
                start = mid
            
        if A[start] == target:
            return start
        if A[end] == target:
            return end
        return -1


    def binary_search_end(self, A, target):
        if A == []:
            return -1
            
        start = 0
        end = len(A)-1
        
        while start+1 < end:
            mid = (start+end) // 2
            
            if A[mid] > target:
                end = mid
            elif A[mid] <= target:
                start = mid
                
        if A[end] == target:
            return end            
        if A[start] == target:
            return start

        return -1
    
    
    def totalOccurrence(self, A, target):
        front_position = self.binary_search_front(A, target)
        end_position = self.binary_search_end(A, target)
        
        if front_position == -1 and end_position == -1:
            return 0
        return end_position-front_position+1
```
