```python
class Solution:
    """
    @param: numCourses: a total of n courses
    @param: prerequisites: a list of prerequisite pairs
    @return: true if can finish all courses or false
    """

    def degree(self, prerequisites):
        course_degree = {}
    
        for course, pre_course in prerequisites:
            if course not in course_degree:
                course_degree[course] = 0
            if pre_course not in course_degree:
                course_degree[pre_course] = 0
            
        for course, pre_course in prerequisites:
            course_degree[course] += 1
            
        return course_degree
        
    
    def make_course_mapping(self, prerequisites):
        course_mapping = {}
    
        for course, pre_course in prerequisites:
            if course not in course_mapping:
                course_mapping[course] = []
            if pre_course not in course_mapping:
                course_mapping[pre_course] = []
        
        for course, pre_course in prerequisites:
            course_mapping[pre_course].append(course)
            
        return course_mapping
        
    
    def canFinish(self, numCourses, prerequisites):            
        course_degree = self.degree(prerequisites)
        course_mapping = self.make_course_mapping(prerequisites)

        start = [int(pre_course) for course, pre_course in prerequisites if course_degree[pre_course] == 0]
        if start == []:
            return False
        else:
            queue = start      
        order = []

        while queue:
            current = queue.pop(0)
            order.append(current)
        
            following_course = course_mapping[current]
            for each_following_course in following_course:
                course_degree[each_following_course] -= 1
                if course_degree[each_following_course] == 0:
                    queue.append(each_following_course)
                
        if len(order) != len(course_mapping):
            return False
        else:
            return True
```
