```python
def n_queen(board_size):
    queen_pos, result = [], []
    search(board_size, queen_pos, result)
    return result
    
    
def search(board_size, queen_pos, result):
    row = len(queen_pos)
    if row == board_size:
        result.append(put_om_board(queen_pos))
        return
    
    for col in range(board_size):
        if not is_valid(col, row, queen_pos):
            continue
        queen_pos.append(col)
        search(board_size, queen_pos, result)
        queen_pos.pop()
        
    
def is_valid(col, row, queen_pos):
    # check column
    for pre_queen_col in queen_pos:
        if pre_queen_col == col:
            return False
    
    # check diagonal
    for i in range(len(queen_pos)):
        pre_queen_row = i
        pre_queen_col = queen_pos[i]
        
        if (abs(col - pre_queen_col) - abs(row - pre_queen_row)) == 0:
            return False
    return True
    
    
def put_om_board(alist):
    board = []
    
    row_length = len(alist)
    col_length = max(alist) + 1
    
    for each_row in alist:
        row_map = ['.'] * col_length
        row_map[each_row] = 'Q'
        board.append(row_map)
        
    return board  
```
